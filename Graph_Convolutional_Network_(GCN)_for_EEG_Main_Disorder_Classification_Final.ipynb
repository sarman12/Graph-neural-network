```python
# @title #  Necessary libraries

import pandas as pd
from tqdm import tqdm
import numpy as np
import torch
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, LabelEncoder
from torch_geometric.data import Data
from torch_geometric.nn import GCNConv
from torch.nn import functional as F
from sklearn.metrics import classification_report
```

```python
# @title # Reducing Feature Space part 1
numeric_data = data.select_dtypes(include=[np.number])

plt.figure(figsize=(12, 8))
corr_matrix = numeric_data.corr()
sns.heatmap(corr_matrix, annot=False, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
# @title # Reducing Feature Space part 2
upper_tri = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
to_drop = [column for column in upper_tri.columns if any(upper_tri[column] > 0.95)]
print(f"Before dropping co-related columns, dataset size is: {len(data.columns)}")
data = data.drop(columns=to_drop)
print(f"Dropped columns: {to_drop}")
print(f"Size of total Dropped columns: {len(to_drop)}")
print(f"After dropping co-related columns, dataset size is: {len(data.columns)}")
```

```python
# @title # Node and Edge features
Nodes = [col for col in data.columns if 'AB' in col]
Edges = [col for col in data.columns if 'COH' in col]

print("Nodes Columns:", Nodes)
print("Edge Columns:", Edges)
print("Total Nodes:", len(Nodes))
print("Total Edges:", len(Edges))
```

```python
# @title # Standardize the features
scaler = StandardScaler()
data[Nodes] = scaler.fit_transform(data[Nodes])
```

```python
X = data[Nodes].values
print("Node Features Shape:", X.shape)
```

```python

coherence_matrix = data[Edges].values
print("Edge shape:",coherence_matrix.shape)
num_nodes = X.shape[1]
num_edges = coherence_matrix.shape[1]
print("Number of Nodes:", num_nodes)
print("Number of Edges:", num_edges)
```

```python
# Define `edge_index` based on coherence matrix structure
edge_list = []
for i in range(num_edges):
    source, target = divmod(i, num_nodes)  # Adjust indices based on your data structure if necessary
    edge_list.append([source, target])
```

```python

edge_array = np.array(edge_list).T
edge_index = torch.tensor(edge_array, dtype=torch.long)
edge_attr = torch.tensor(coherence_matrix.flatten(), dtype=torch.float)
```

```python
# @title #Creating Graph Object
x = torch.tensor(X, dtype=torch.float)
y = torch.tensor(data['main.disorder'].values, dtype=torch.long)
graph_data = Data(x=x, edge_index=edge_index, edge_attr=edge_attr, y=y)
```

```python
# @title # Graph information

print("Graph Information:")
print(f"Number of Nodes: {graph_data.num_nodes}")
print(f"Number of Edges: {graph_data.num_edges}")
print(f"Node Features Shape: {graph_data.x.shape}")
print(f"Edge Index Shape: {graph_data.edge_index.shape}")
print(f"Edge Attributes Shape: {graph_data.edge_attr.shape if graph_data.edge_attr is not None else 'None'}")
print(f"Labels Shape: {graph_data.y.shape}")
```

```python
# @title #Initialize model, optimizer, and loss function

def init_model():
    model = GCNMultiClass()
    optimizer = torch.optim.Adam(model.parameters(), lr=0.009)
    criterion = torch.nn.CrossEntropyLoss()
    return model, optimizer, criterion

def train(model, optimizer, criterion):
    model.train()
    optimizer.zero_grad()
    out = model(graph_data)
    loss = criterion(out, graph_data.y)
    loss.backward()
    optimizer.step()
    return loss.item()

def evaluate(model):
    model.eval()
    _, pred = model(graph_data).max(dim=1)
    correct = int(pred.eq(graph_data.y).sum().item())
    accuracy = correct / len(graph_data.y)
    return accuracy
```

```python
# @title #Accuracies
accuracies = []
all_epoch_losses = []
all_epoch_accuracies = []

for run in range(50):
    model, optimizer, criterion = init_model()
    epoch_losses = []
    epoch_accuracies = []

    for epoch in tqdm(range(1, 201), desc=f"Run {run+1} Progress", leave=False):
        loss = train(model, optimizer, criterion)
        epoch_losses.append(loss)

        if epoch % 10 == 0:
            acc = evaluate(model)
            epoch_accuracies.append(acc)
    accuracies.append(evaluate(model))
    all_epoch_losses.append(epoch_losses)
    all_epoch_accuracies.append(epoch_accuracies)
    print(f'Run {run+1}, Final Accuracy: {accuracies[-1]:.4f}')
```

```python
average_accuracy = sum(accuracies) / len(accuracies)
print(f'Average Accuracy over 50 runs: {average_accuracy:.4f}')

average_epoch_losses = np.mean(all_epoch_losses, axis=0)
average_epoch_accuracies = np.mean(all_epoch_accuracies, axis=0)

plt.figure(figsize=(14, 5))

plt.subplot(1, 2, 1)
plt.plot(average_epoch_losses, label='Average Loss over 50 Runs', color='blue')
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Training Loss over Epochs")
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(range(10, 201, 10), average_epoch_accuracies, label='Average Accuracy over 50 Runs', color='green')
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.title("Training Accuracy over Epochs")
plt.legend()

plt.tight_layout()
plt.show()
```
